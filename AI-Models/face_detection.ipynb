# face detection through the camera

```python
import cv2
import matplotlib.pyplot as plt


# Start the laptop's camera
cap = cv2.VideoCapture(0)  # '0' is typically the default value for the primary camera

# Load the pre-trained Haar Cascade model for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Check if the camera opened successfully
if not cap.isOpened():
    print("Cannot open camera")
else:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if frame is received
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
    else:
        # Convert the captured frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Detect faces in the grayscale frame
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # Draw rectangles around the faces
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)  # Draw a green rectangle with thickness 2

        # Convert BGR image to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Display the resulting frame with face corners
        plt.imshow(frame_rgb)
        plt.axis('off')  # Hide axis
        plt.show()

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

# crop the detected face 

```python
import cv2
import matplotlib.pyplot as plt

# Start the laptop's camera
cap = cv2.VideoCapture(0)  # '0' is typically the default value for the primary camera

# Load the pre-trained Haar Cascade model for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Check if the camera opened successfully
if not cap.isOpened():
    print("Cannot open camera")
else:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if frame is received
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
    else:
        # Convert the captured frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Detect faces in the grayscale frame
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # List to hold cropped faces
        cropped_faces = []

        # Draw rectangles around the faces and crop the faces
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)  # Draw a green rectangle with thickness 2
            # Crop the face
            cropped_face = frame[y:y+h, x:x+w]
            cropped_faces.append(cropped_face)

        # Convert BGR image to RGB for full image display
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Display the original frame with rectangles
        plt.figure(figsize=(10, 6))
        plt.imshow(frame_rgb)
        plt.axis('off')  # Hide axis
        plt.title("Detected Faces")
        plt.show()

        # Display each cropped face
        for i, face in enumerate(cropped_faces):
            plt.figure(figsize=(2, 2))  # Size can be adjusted based on your preference
            face_rgb = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
            plt.imshow(face_rgb)
            plt.axis('off')
            plt.title(f"Face {i+1}")
            plt.show()

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

# set dimensions and convert it to greyscale and from RGBA to RGB

```python
import cv2
import matplotlib.pyplot as plt

# Start the laptop's camera
cap = cv2.VideoCapture(0)  # '0' is typically the default value for the primary camera

# Load the pre-trained Haar Cascade model for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Check if the camera opened successfully
if not cap.isOpened():
    print("Cannot open camera")
else:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if frame is received
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
    else:
        # Convert the captured frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Detect faces in the grayscale frame
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # List to hold cropped, resized, and processed faces
        processed_faces = []

        # Draw rectangles around the faces, crop, resize, and process the faces
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)  # Draw a green rectangle with thickness 2
            # Crop the face from the grayscale image
            cropped_face = gray[y:y+h, x:x+w]
            # Resize the cropped face
            resized_face = cv2.resize(cropped_face, (320, 243))
            # Convert grayscale to RGB (for display purposes)
            face_rgb = cv2.cvtColor(resized_face, cv2.COLOR_GRAY2RGB)
            processed_faces.append(face_rgb)

        # Convert BGR image to RGB for full image display
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Display the original frame with rectangles
        plt.figure(figsize=(10, 6))
        plt.imshow(frame_rgb)
        plt.axis('off')  # Hide axis
        plt.title("Detected Faces")
        plt.show()

        # Display each processed face
        for i, face in enumerate(processed_faces):
            plt.figure(figsize=(4, 3))  # Size adjusted to match aspect ratio of the resized images
            plt.imshow(face)
            plt.axis('off')
            plt.title(f"Face {i+1}")
            plt.show()

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
